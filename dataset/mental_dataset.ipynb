```python
import pandas as pd
import numpy as np
import random
```

```python
Diagnose = ['Depression', 'Stress', 'Anxiety']
Depression_mood =['Sad', 'Empty', 'Hopeless', 'Normal'] 
Diminished_pleasure_interest = [True, False]
Weight= ['Gain', 'Loss', 'Normal']
Appetite = ['Increase', 'Decrease', 'Normal']
Sleeping_disorder = ['Insomnia', 'hypersomnia', 'Normal']
Psychomotor_agitation_retardation = ['Restlessness', 'Slowed down', 'Normal']
Fatigue_losses_energy = [True, False]
Worthlessness_inappropriate_guilt = [True, False]
Thinking_concentrate = ['Indecisiveness', 'Decisiveness']
Suicidal_idea = [True, False]
Intrusion_symptoms = ['Intrusive distressing memories', 'Recurrent distressing dreams', 'Flashbacks', 'Prolonged psychological distress to internal and external cues', 'Prolonged physiological reactions to internal and external cues', 'Normal']
Negative_mood = ['Inability to experience positive emotions', 'No happiness', 'No satisfaction', 'No loving feelings', 'Normal']
Dissociative_symptoms = ['Altered sense of reality of oneself or surrounding', 'Inability to remember traumatic events', 'Head injury', 'Alcohol', 'Drugs', 'Normal']
Avoidance_symptoms = ['Avoid distressing memories', 'Avoid distressing thoughts', 'Avoid distressing feelings', 'Avoid external reminders', 'Normal']
Arousal_symptoms = ['Sleep disturbance', 'Irritable behavior', 'Angry outbursts', 'Hypervigilance', 'Problem with concentration', 'Exaggrated startle response', 'Normal']
Restlessness_feeling_keyed_up_on_edge = [True, False]
Being_easily_fatigued = [True, False]
Difficult_concentrate_mind_going_blank = [True, False]
Irritability = ['High percentage', 'Low percentage']
Muscle_tension = ['Unrelax', 'Relax']
Sleep_disturbance = ['Difficult falling asleep', 'Difficult staying asleep', 'Unsatisfying sleep', 'Normal']
Worry = ['High percentage', 'Low percentage']
```

```python
dataset = {'depression mood' :[], 'diminished pleasure/interest' :[], 'weight' :[], 'appetite' :[], 'sleeping disorder' :[], 'psychomotor agitation/retardation' :[], 'fatigue/energy loss' :[], 'worthlessness/inappropriate guilt' :[], 'thinking abilty/concentration' :[], 'suicidal idea' :[], 'intrusion symptoms':[], 'negative mood' :[], 'dissociative symptoms' :[], 'avoidance symptoms' :[], 'arousal symptoms' :[], 'restlessness/feeling keyed up or on edge' :[], 'easily fatigued' :[], 'difficult concentrating/mind going blank' :[], 'irritability' :[], 'muscle tension' :[], 'sleep disturbance' :[], 'worry' :[], 'diagnose' :[]}
```

```python
for i in range(0, 5000):
    factors = []
    stress_factors = []
    anxiety_factors = []
    Depression_mood_data= random.choice(Depression_mood)
    dataset['depression mood'].append(Depression_mood_data)
    Diminished_pleasure_interest_data= random.choice(Diminished_pleasure_interest)
    dataset['diminished pleasure/interest'].append(Diminished_pleasure_interest_data)
    Weight_data= random.choice(Weight)
    dataset['weight'].append(Weight_data)
    Appetite_data= random.choice(Appetite)
    dataset['appetite'].append(Appetite_data)
    Sleeping_disorder_data= random.choice(Sleeping_disorder)
    dataset['sleeping disorder'].append(Sleeping_disorder_data)
    Psychomotor_agitation_retardation_data= random.choice(Psychomotor_agitation_retardation)
    dataset['psychomotor agitation/retardation'].append(Psychomotor_agitation_retardation_data)
    Fatigue_losses_energy_data= random.choice(Fatigue_losses_energy)
    dataset['fatigue/energy loss'].append(Fatigue_losses_energy_data)
    Worthlessness_inappropriate_guilt_data= random.choice(Worthlessness_inappropriate_guilt)
    dataset['worthlessness/inappropriate guilt'].append(Worthlessness_inappropriate_guilt_data)
    Thinking_concentrate_data= random.choice(Thinking_concentrate)
    dataset['thinking abilty/concentration'].append(Thinking_concentrate_data)
    Suicidal_idea_data= random.choice(Suicidal_idea)
    dataset['suicidal idea'].append(Suicidal_idea_data)
    Intrusion_symptoms_data= random.choice(Intrusion_symptoms)
    dataset['intrusion symptoms'].append(Intrusion_symptoms_data)
    Negative_mood_data= random.choice(Negative_mood)
    dataset['negative mood'].append(Negative_mood_data)
    Dissociative_symptoms_data= random.choice(Dissociative_symptoms)
    dataset['dissociative symptoms'].append(Dissociative_symptoms_data)
    Avoidance_symptoms_data= random.choice(Avoidance_symptoms)
    dataset['avoidance symptoms'].append(Avoidance_symptoms_data)
    Arousal_symptoms_data= random.choice(Arousal_symptoms)
    dataset['arousal symptoms'].append(Arousal_symptoms_data)
    Restlessness_feeling_keyed_up_on_edge_data= random.choice(Restlessness_feeling_keyed_up_on_edge)
    dataset['restlessness/feeling keyed up or on edge'].append(Restlessness_feeling_keyed_up_on_edge_data)
    Being_easily_fatigued_data= random.choice(Being_easily_fatigued)
    dataset['easily fatigued'].append(Being_easily_fatigued_data)
    Difficult_concentrate_mind_going_blank_data= random.choice(Difficult_concentrate_mind_going_blank)
    dataset['difficult concentrating/mind going blank'].append(Difficult_concentrate_mind_going_blank_data)
    Irritability_data= random.choice(Irritability)
    dataset['irritability'].append(Irritability_data)
    Muscle_tension_data= random.choice(Muscle_tension)
    dataset['muscle tension'].append(Muscle_tension_data)
    Sleep_disturbance_data= random.choice(Sleep_disturbance)
    dataset['sleep disturbance'].append(Sleep_disturbance_data)
    Worry_data= random.choice(Worry)
    dataset['worry'].append(Worry_data)
    Diagnose_data =''
    if Depression_mood_data != 'Normal':
        factors.append(1)
    if Diminished_pleasure_interest_data != False:
        factors.append(1)
    if Weight_data != 'Normal':
        factors.append(1)
    if Appetite_data != 'Normal':
        factors.append(1)
    if Sleeping_disorder_data != 'Normal':
        factors.append(1)
    if Psychomotor_agitation_retardation_data != 'Normal':
        factors.append(1)
    if Fatigue_losses_energy_data != False:
        factors.append(1)
    if Worthlessness_inappropriate_guilt_data != False:
        factors.append(1)
    if Thinking_concentrate_data != 'Decisiveness':
        factors.append(1)
    if Suicidal_idea_data != False:
        factors.append(1)
    if Depression_mood_data != 'Normal' or Diminished_pleasure_interest_data != False:
        Diagnose_data +='Depression '
    else:
        if len(factors)>=5:
            Diagnose_data +='Depression '
        else:
          Diagnose_data +='Normal ' 
    if Intrusion_symptoms_data != 'Normal':
        stress_factors.append(1)
    if Negative_mood_data != 'Normal':
        stress_factors.append(1)
    if Dissociative_symptoms_data != 'Normal':
        stress_factors.append(1)
    if Avoidance_symptoms_data != 'Normal':
        stress_factors.append(1)
    if Arousal_symptoms_data != 'Normal':
        stress_factors.append(1)
    if Intrusion_symptoms_data != 'Normal' or Negative_mood_data != 'Normal':
        Diagnose_data +='Stress '
    else:
        if len(stress_factors)>=5:
            Diagnose_data +='Stress '
        else:
          Diagnose_data +='Normal '
    if Restlessness_feeling_keyed_up_on_edge_data != False:
        anxiety_factors.append(1)
    if Being_easily_fatigued_data != False:
        anxiety_factors.append(1)
    if Difficult_concentrate_mind_going_blank_data != False:
        anxiety_factors.append(1)
    if Irritability_data != 'Low percantage':
        anxiety_factors.append(1)
    if Muscle_tension_data != 'Relax':
        anxiety_factors.append(1)
    if Sleep_disturbance_data != 'Normal':
        anxiety_factors.append(1)
    if Worry_data != 'Low percantage':
        anxiety_factors.append(1)
    if Restlessness_feeling_keyed_up_on_edge_data != False :
        Diagnose_data +='Anxiety '
    else:
        if len(stress_factors)>=3:
            Diagnose_data +='Anxiety '
        else:
          Diagnose_data +='Normal '
        
    dataset['diagnose'].append(Diagnose_data)
```

```python
data = pd.DataFrame(dataset)
```
